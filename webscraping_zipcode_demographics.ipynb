The column 'view' tells us how many times the house has been viewed before purchase. With only 63 rows having a null value for the views column we can choose to drop it (because it is only 0.2% of the data) or we can choose to impute a measure of central tendency.

```python
df.waterfront.fillna(0, inplace=True)
```

```python
df.yr_renovated.fillna(0, inplace=True)
```

Web-Scraping

```python
zipcodes = list(df['zipcode'].unique())
```

```python
from bs4 import BeautifulSoup
import requests
import re

# Permissions to access webpage, use "headers=headers" with requests
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) '\
           'AppleWebKit/537.36 (KHTML, like Gecko) '\
           'Chrome/75.0.3770.80 Safari/537.36'}

# Making Soup to Extract Demographics
url = {}
response = {}
soup = {}
demographics = {}
for x in zipcodes:
    url[x] = ('https://www.unitedstateszipcodes.org/%s/' % x)
    response[x] = requests.get(url[x], headers = headers)
    soup[x] = BeautifulSoup(response[x].text, 'lxml')
    # Container
    demographics[x] = soup[x].find_all('div', class_='col-xs-12 col-sm-6')
```

```python
# Testing Cell

ex1 = 98033
ex2 = 98155
print(soup[ex2])
#demographics[example][0].find_all('td', class_='text-right')[0].get_text()
```

```python
# Scraping Demographics
pop = []
sqmile = []
popdensity = []
units = []
units_occupied = []
med_homevalue = []
med_householdincome = []

for y in zipcodes:
    try:
        pop.append(demographics[y][0].find_all('td', class_='text-right')[0].get_text())
        sqmile.append(demographics[y][1].find_all('td', class_='text-right')[0].get_text())
        popdensity.append(demographics[y][0].find_all('td', class_='text-right')[1].get_text())
        units.append(demographics[y][0].find_all('td', class_='text-right')[2].get_text())
        units_occupied.append(demographics[y][1].find_all('td', class_='text-right')[2].get_text())
        med_homevalue.append(demographics[y][0].find_all('td', class_='text-right')[3].get_text())
        med_householdincome.append(demographics[y][1].find_all('td', class_='text-right')[3].get_text())
    except:
        pop.append(0)
        sqmile.append(0)
        popdensity.append(0)
        units.append(0)
        units_occupied.append(0)
        med_homevalue.append(0)
        med_householdincome.append(0)
```

```python
# Creating DataFrame

zipcode_df = pd.DataFrame([zipcodes, pop, sqmile, popdensity, units, units_occupied, med_homevalue, med_householdincome]).transpose()
zipcode_df.columns = ['zipcode', 'population', 'land', 'population_density', 'units', 'units_occupied', 'med_value', 'med_income']
zipcode_df.head(70)
```
