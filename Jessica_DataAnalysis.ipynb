### Question: How does the population density in King County affect housing prices? 

When there is a demand for real estate  In order to see how _____________ affects housing prices we'll look at the average housing price within a certain interval. Each interval.

```python
sns.relplot(x="calc_pd", y="price", hue="calc_pd",
            sizes=(40, 400), alpha=.5, palette="muted", #size="weight",
            height=6, data=df)
```

```python
df = df[df.price < 5500000]
```

Dividing by categories

```python
labels = ["{0} - {1}".format(i, i + 3200) for i in range(0, 16000, 3200)]
df['Population_Density'] = pd.cut(df.calc_pd, range(0, 16001, 3200), right=False, labels=labels)
```

```python
plt.figure(figsize=(8,5))
ax_count = sns.countplot(x="Population_Density", data=df,
                         palette="Blues_d")
ax_count.set_title('Number of Houses Sold within Population Density Range')
```

Above shows counts

```python
plt.figure(figsize=(8,5))
ax = sns.barplot(x="Population_Density", y="price", data=df,
                 linewidth=2.5, facecolor="salmon",
                 errcolor=".2", edgecolor=".2")
ax.set_title('Average House Prices within Population Density Range')
```

Above shows average prices
