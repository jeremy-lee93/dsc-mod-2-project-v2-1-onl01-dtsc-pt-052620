```python
import pandas as pd
```

```python
df = pd.read_csv('DataTable.csv')
df.info()
```

```python
df.drop(columns=['id', 'sqft_above', 'sqft_basement', 'zipcode', 'population', 'land', 'population_density', 'view', 'grade', 'distance_to_seahawks_stadium', 'distance_to_kerry_park', 'distance_to_discovery_park', 'units', 'units_occupied', 'condition'])
```

```python
import seaborn as sns
```

```python
df['bxb'] = df['bedrooms']*df['bathrooms']
g = sns.PairGrid(df, vars=['price', 'bedrooms', 'bathrooms', 'bxb'])
g.map(sns.scatterplot)
```

```python
df['live:lot'] = df['sqft_living']/df['sqft_lot']
g = sns.PairGrid(df, vars=['price', 'sqft_living', 'sqft_lot','live:lot'])
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, vars=['price', 'floors', 'waterfront'])
g.map(sns.scatterplot)
```

```python
df['yr_today'] = 0
if (df['yr_renovated'] == 0):
    df['yr_today'] = 2020-df['yr_built']
else: 
    df['yr_today'] = 2020-df['yr_renovated']
g = sns.PairGrid(df, vars=['price', 'yr_built', 'yr_renovated', 'yr_today'])
g.map(sns.scatterplot)
```

```python
df['lxl'] = df['lat']*df['long']
g = sns.PairGrid(df, vars=['price', 'lat', 'long', 'lxl'])
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, vars=['price', 'distance_to_airport', 'distance_to_Seattle_center', 'minimum_distance_to_metro'])
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(df, vars=['price', 'med_value', 'med_income', 'calc_pd'])
g.map(sns.scatterplot)
```
